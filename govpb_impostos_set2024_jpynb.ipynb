```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/matriz_impostos.xlsx')
```

```python
df
```

```python
# 0. FILTRAR COLUNA POR ORDEM ALFABÉTICA DECRESCENTE
order_alphab = df.sort_values(by='Veículo', ascending=False)
order_alphab
```

```python
# 1. FILTRAR COLUNAS DESEJADAS
column_filter = df[['Veículo','Contratado', 'Entregues']]
column_filter
```

```python
# 2. REMOVER ESPAÇOS EM BRANCO E NORMALIZAR A FORMATAÇÃO DOS DADOS DA COLUNA
dados_filter = df['Veículo'].str.strip().str.capitalize()
# REMOVENDO OS NÚMEROS ENTRE PARENTESES INSERIDOS NAS COLUNAS
remove_number_dados = dados_filter.str.replace(r'\(\d+\)', '', regex=True)
remove_number_dados
veiculos_totais=df[['Veículo','Contratado', 'Entregues']]
veiculos_totais
```

```python
# VERIFICAR SE EXISTEM DADOS DUPLICADOS NA COLUNA VEÍCULO
dados = remove_number_dados
dados[dados.duplicated()]
```

```python
dados.value_counts()
```
